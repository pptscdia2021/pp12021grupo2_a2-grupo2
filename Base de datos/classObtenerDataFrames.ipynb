```python
import pymysql
import pandas as pd

class obtenerDataFrames():
    def __init__(self):
        pass
    
    def otenerDatos(self, db, name, consulta):
        self.db = db
        conexion = pymysql.connect(host='localhost',
                             user='root',
                             password='',
                             db = db)
        
        self.consulta = consulta
        self.name = name
        name = consulta
        name = pd.read_sql_query(name,conexion)
        return name.head()
```

```python
empleados_totales = obtenerDataFrames()
empleados_totales.otenerDatos('employees','empleados','SELECT * FROM employees')
```

```python
empleadas_mujeres = obtenerDataFrames()
empleadas_mujeres.otenerDatos('employees','empleadas',"SELECT * FROM employees WHERE gender ='F'")
```

```python
departamentos = obtenerDataFrames()
departamentos.otenerDatos('employees','departamentos','SELECT * FROM departments')
```
