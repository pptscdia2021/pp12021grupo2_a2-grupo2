```python
import pandas as pd
import requests
import yfinance as yf
import investpy
from bs4 import BeautifulSoup
from pymongo import MongoClient
import matplotlib.pyplot as plt
from datetime import datetime
from os import remove
from bson.json_util import dumps, loads
```

```python
df = pd.read_csv('yahooHistoricos.csv')
```

```python
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(30, 30))
fig.suptitle('Datos históricos acciones YahooFinance')
for i in range(len(df)):
    axe = plt.subplot(7, 5, i+1)
    axe.set_title(df['symbol'][i])
    df[i].plot(kind='line', y='ult', x='fecha', ax=axe)
fig.tight_layout(pad=2.0)
```
